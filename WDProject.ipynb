```python
###### Create a Unit Box ######
import matplotlib.pyplot as plt
import numpy as np

def unitbox(l, h):    
    x = np.arange(-10, 10) 
    y = np.zeros(len(x))+h
    ynew = np.where(np.absolute(x) > l/2, y, 0) 
    return x, ynew

testbox = unitbox(2,3)
print testbox

plt.plot(testbox[0], testbox[1])
plt.xlabel('Time')
plt.ylabel('Flux')
plt.title('Transit Unit Box')
plt.axis([-5, 5, -1, 6])
plt.grid(True)
plt.show()
```
